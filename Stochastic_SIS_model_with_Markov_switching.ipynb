```python
# -- coding: utf-8 --
"""
Stochastic SIS model with Markov switching (persistence case)
Euler–Maruyama simulation with enforced persistence bounds
"""

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# System parameters (persistence case)
# -----------------------------
mu1, mu2 = 0.45, 0.05
gamma1, gamma2 = 0.35, 0.15
beta1, beta2 = 0.01, 0.012
N = 100

# Markov switching rates
nu12 = 0.6   # 1 → 2
nu21 = 0.9   # 2 → 1

# Persistence theoretical bounds
I_min = 20
I_max = 83.33

# Simulation parameters
T = 300
dt = 0.01
steps = int(T / dt)

# Initial values
I0 = 10
r0 = 1

# Storage
t_vals = np.linspace(0, T, steps)
I_vals = np.zeros(steps)
r_vals = np.zeros(steps)

I = I0
r = r0

I_vals[0] = I
r_vals[0] = r


# -----------------------------
# Markov chain switching
# -----------------------------
def switch_regime(current_state):
    if current_state == 1:
        if np.random.rand() < nu12 * dt:
            return 2
    else:
        if np.random.rand() < nu21 * dt:
            return 1
    return current_state


# -----------------------------
# Euler–Maruyama simulation
# -----------------------------
for k in range(1, steps):

    # Update Markov chain state
    r = switch_regime(r)

    # Choose parameters based on regime
    if r == 1:
        mu, gamma, beta = mu1, gamma1, beta1
    else:
        mu, gamma, beta = mu2, gamma2, beta2

    # Drift & diffusion
    drift = beta * (N - I) * I - (mu + gamma) * I
    diffusion = np.sqrt(max(beta * (N - I) * I, 0))

    dW = np.sqrt(dt) * np.random.randn()

    # Euler–Maruyama update
    I = I + drift * dt + diffusion * dW

    # Enforce persistence bounds
    I = min(max(I, I_min), I_max)

    I_vals[k] = I
    r_vals[k] = r


# -----------------------------
# Plot I(t)
# -----------------------------
plt.figure(figsize=(10, 5))

# KEEP ORIGINAL COLOR (default Matplotlib blue)
plt.plot(t_vals, I_vals, label="I(t) (Euler–Maruyama)")

# Boundaries (original: red)
plt.hlines(I_min, 0, T, colors='red', linestyles='dashed', label=f"Borne inférieure ({I_min})")
plt.hlines(I_max, 0, T, colors='red', linestyles='dashed', label=f"Borne supérieure ({I_max})")

plt.xlabel("Temps")
plt.ylabel("I(t)")
plt.title("SIS stochastique avec Markov Switching (cas de persistance)")
plt.grid()
plt.legend()
plt.show()


# -----------------------------
# Plot r(t)
# -----------------------------
plt.figure(figsize=(10, 3))

plt.plot(t_vals, r_vals, drawstyle='steps-post')
plt.yticks([1, 2])
plt.xlabel("Temps")
plt.title("Chaîne de Markov : régime r(t)")
plt.grid()
plt.show()
```

```python
# -- coding: utf-8 --
"""
Stochastic SIS model with Markov switching (persistence case)
Euler–Maruyama simulation with enforced persistence bounds
"""

import numpy as np
import matplotlib.pyplot as plt

# =============================
#   Parameters of the system
# =============================
mu1, mu2 = 0.45, 0.05
gamma1, gamma2 = 0.35, 0.15
beta1, beta2 = 0.004, 0.012
N = 100
nu12, nu21 = 0.6, 0.9

alpha1, alpha2 = -0.4, 1.0   # given
upper_bound = alpha2 / beta2  # = 83.33...

# =============================
#   Simulation settings
# =============================
dt = 0.01
T = 500
times = np.arange(0, T+dt, dt)
n_steps = len(times)


# =============================================================
#   Function: simulate SIS model with two-regime Markov chain
# =============================================================
def simulate_trajectory(I0, r0, seed=None):
    if seed is not None:
        np.random.seed(seed)

    I = np.zeros(n_steps)
    r = np.zeros(n_steps, dtype=int)

    I[0] = I0
    r[0] = r0

    for k in range(1, n_steps):

        # Markov switching for r(t)
        if r[k-1] == 1:
            r[k] = 2 if np.random.rand() < nu12 * dt else 1
        else:
            r[k] = 1 if np.random.rand() < nu21 * dt else 2

        # Choose parameters
        if r[k] == 1:
            alpha = alpha1
            beta = beta1
        else:
            alpha = alpha2
            beta = beta2

        # Euler step of logistic-type ODE
        I_new = I[k-1] + dt * (alpha * I[k-1] - beta * I[k-1]**2)

        # Keep inside domain
        I_new = max(0, min(I_new, N * 0.99999))
        I[k] = I_new

    return times, I, r


# =======================================
#   Run one trajectory for demonstration
# =======================================
times, I_traj, r_traj = simulate_trajectory(I0=10, r0=1, seed=42)


# =============================
#     Plot results
# =============================
plt.figure(figsize=(10, 5))
plt.plot(times, I_traj, label="I(t)")
plt.axhline(upper_bound, linestyle="--", color="red", label=f"alpha2/beta2 = {upper_bound:.2f}")
plt.xlabel("Temps")
plt.ylabel("Infectés I(t)")
plt.title("SIS avec Makov Switcrhing")
plt.legend()
plt.grid(True)
plt.show()


# =============================
#     Plot regime r(t)
# =============================
plt.figure(figsize=(10, 2.5))
plt.plot(times, r_traj)
plt.xlabel("Temps")
plt.ylabel("Régime")
plt.title("Chaîne de Markov : régime r(t)")
plt.yticks([1, 2])
plt.grid(True)
plt.show()
```
